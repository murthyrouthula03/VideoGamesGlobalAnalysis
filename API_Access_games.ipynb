```python
import requests
import pandas as pd
import numpy as np
import json
import os
```

```python
# Accessing API of All Video games
url_games = "http://127.0.0.1:5000/api/v1/Videogames/all.json"
data_games = requests.get(url_games)
```

```python
# checking response
print(data_games)
```

```python
# getting data in json format
data_games_info = data_games.json()
print(data_games_info)
```

```python
# converting JSON data to pandas dataframe
game_df = pd.DataFrame(data_games_info)
game_df
```

```python
# getting API data based on the requirement through query
filter_url = "http://127.0.0.1:5000/api/v1/Videogames/all.json?$select=0"
filter_data = requests.get(filter_url).json()
print(filter_data)
```
